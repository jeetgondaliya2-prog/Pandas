```python
import numpy as np             # filtering dataframe on condition
import pandas as pd 
```

```python
data = pd.read_csv('matches.csv')
```

```python
data['city']=='Delhi'        # check city  (boolean series)
```

```python
mask= data['city']=='Delhi'  # data when city==Delhi &  by data[mask].shape it give total row and column where row denote total counts
data[mask]
```

```python
def get_city_match_count(city):         # function to count the match count in different city 
    mask=data['city']==city
    return data[mask].shape[0]
```

```python
get_city_match_count('Delhi')
```

```python
get_city_match_count('Bangalore')
```
