```python
import numpy as np              #pivot table 
import pandas as pd 
```

```python
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

```python
mask=delivery['batsman_runs']==6
```

```python
six=delivery[mask]                    # total six hit in ipl
six.shape[0]
```

```python
six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')            #pivot table
```

```python
pt=six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
```
