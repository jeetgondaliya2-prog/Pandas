```python
import numpy as np                 # isin function
import pandas as pd
```

```python
delivery=pd.read_csv('deliveries.csv')
```

```python
# find the most destructive death over batsman in history of ipl
# strike rate = (no of run / no of balls)*100
# batsman plays min 200 balls in 16-20 overs
```

```python
delivery.groupby('batsman')['batsman_runs'].count()   # balls played by batsman
```

```python
death_over=delivery[delivery['over']>15]
```

```python
death_over.groupby('batsman')['batsman_runs'].count()             # batsman play  ball in 16-20 overs
```

```python
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()     # batsman play more than 200 balls in 16-20 overs
x=all_batsman>200
all_batsman[x]
```

```python
batsman_list=all_batsman[x].index.tolist()
```

```python
# runs and balls played by all these 43 batsman
```

```python
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
sr.sort_values(ascending=False).head(1)                       # most destructive batsman
```
